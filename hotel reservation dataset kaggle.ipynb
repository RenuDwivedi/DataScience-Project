```python
df.drop(columns=['Booking_ID',],inplace=True)
df
```

```python
df=pd.get_dummies(df,drop_first=True)
df
```

```python
x= df.drop(columns = 'booking_status_Not_Canceled')
y =df['booking_status_Not_Canceled']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
 
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
ypred_train = model.predict(x_train)
ypred_test = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,plot_confusion_matrix,classification_report
```

```python
print(accuracy_score(y_train,ypred_train))
print(accuracy_score(y_test,ypred_test))
```

```python
print(confusion_matrix(y_test,ypred_test))
print(plot_confusion_matrix(model,x_test,y_test))
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(model,x,y,cv=5)
print(score)
score.mean()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,ypred_test))
```
